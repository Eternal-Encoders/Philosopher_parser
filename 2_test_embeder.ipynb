```python
import torch
import pandas as pd
import numpy as np
from sentence_transformers import SentenceTransformer
```

```python
m_path = 'google/embeddinggemma-300m'
```

```python
model = SentenceTransformer(
    m_path,
    model_kwargs={
        'dtype': torch.bfloat16,
        'device_map': 'auto'
    }
)

model = torch.compile(
    model,
    mode='max-autotune',
    fullgraph=True
)
```

```python
df = pd.DataFrame(pd.read_pickle('./__output__/binaries/nodes_data.pkl'))

df
```

```python
embedings = model.encode(
    df['text'].to_list(),
    prompt_name='document',
    batch_size=16,
    show_progress_bar=True,
    normalize_embeddings=True
)
```

```python
search = 'Какие философские концепции встречались в истории?'
search_emb = model.encode(
    search,
    prompt_name='query',
    normalize_embeddings=True
)

sims = embedings @ search_emb
top_k = np.argsort(sims).reshape(-1)[-10:]
df.iloc[top_k]['text'].to_list()
```

```python
search = 'Почему Бэкон и Декарт отвергали идею постепенного развития научного знания?'
search_emb = model.encode(
    search,
    prompt_name='query',
    normalize_embeddings=True
)

sims = embedings @ search_emb
top_k = np.argsort(sims).reshape(-1)[-10:]
df.iloc[top_k]['text'].to_list()
```
